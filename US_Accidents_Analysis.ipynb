### Start Time

```python
df.Start_Time[0] #Format is yyyy-mm-dd
```

```python
df.Start_Time = pd.to_datetime(df.Start_Time)
```

```python
df.Start_Time[0]
```

```python
#Getting the hour from timestamp
hours = df.Start_Time.dt.hour
sns.distplot(hours, bins = 24, kde = False, norm_hist = True) #bins is for 24 hours
```

- High percentage of accidents occur betweeen 3pm and 6pm
- Next highest is between 6am to 10am

```python
days = df.Start_Time.dt.dayofweek
sns.distplot(days, bins = 7, kde = False, norm_hist = True)
```

- Most accidents happen during weekdays
- Amonst weekdays, friday has the highest frequency
- Weekends have lower accidents

### Hour by hour analysis of weekends

```python
sunday_hours = df.Start_Time[df.Start_Time.dt.dayofweek == 6]
saturday_hours = df.Start_Time[df.Start_Time.dt.dayofweek == 5]
```

```python
sns.distplot(sunday_hours.dt.hour, bins = 24, kde = False, norm_hist = True)
```

```python
sns.distplot(saturday_hours.dt.hour, bins = 24, kde = False, norm_hist = True)
```

```python
import folium
from folium import plugins
from folium.plugins import HeatMap
```

```python
folium.Map()
lat, longi = df.Start_Lat[0], df.Start_Lng[0]
folium.Map(location = [lat, longi])
```

```python
#Making a HeatMap
map = folium.Map()
samples = df.sample(int(0.01* len(df)))
lat_long_pairs = zip(list(samples.Start_Lat), list(samples.Start_Lng))
HeatMap(lat_long_pairs).add_to(map)
map
```
